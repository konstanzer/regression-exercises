```python
import pandas as pd
from env import host, username, password
import explore
```

```python
def get_db_url(username, host, password, db):
    return f'mysql+pymysql://{username}:{password}@{host}/{db}'

def get_data():
	url = get_db_url(username, host, password, 'mall_customers')
	query = """
		SELECT *
		FROM customers;
		"""
	return pd.read_sql(query, url)
```

```python
df = get_data()
df.head()
```

```python
explore.plot_variable_pairs(df)
explore.plot_categorical_and_continuous_vars(df, 'gender', 'spending_score')
```
